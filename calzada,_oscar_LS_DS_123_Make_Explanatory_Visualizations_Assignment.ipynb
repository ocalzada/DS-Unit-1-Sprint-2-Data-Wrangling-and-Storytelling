```python
final = last_day_filtered.tail(1)
final.T
```

```python
pct_columns = ['1_pct', '2_pct', '3_pct', '4_pct', '5_pct', '6_pct', '7_pct', '8_pct',
              '9_pct', '10_pct']

final[pct_columns].T
```

```python
plot_data = final[pct_columns].T
plot_data.index = range(1,11)
plot_data
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('fivethirtyeight')

fig = plt.figure()
fig.patch.set(facecolor='white')

ax = plot_data.plot.bar(color='C1', width=0.9)
ax.set(facecolor='white')
ax.text(x=-1.7, y=42, s='IMDb ratings for the films as of Aug 29', fontsize=12)
#ax.set_xticklabels(range(1,11,1))
#ax.set_yticklabels(range(10,50,10))

ax.get_legend().remove()

plt.xlabel('rating', fontweight='bold')
plt.ylabel('Percent of Total votes', weight='bold')
plt.title("'An Incovenient Sequel: Truth to Power' is divisive", fontfamily='sans-serif',
          fontstyle='normal', fontsize=13, fontweight='bold', x=-.1, y=1.1, loc='left')
plt.figtext(0, -0.2, 'FiveThirtyEight', fontsize=10, color='gray', fontweight='normal', wrap=True)
plt.figtext(0.8, -0.2, 'Source: IMDB', fontsize=10, color='gray', fontweight='normal', wrap=True)
             
plt.show()
```

```python
sides_simplified = sides_simplified.fillna(0)
```

```python
sides_simplified['brussel_sprouts'] = sides_simplified['brussel_sprouts'].replace('brussel_sprouts', 1)
sides_simplified['carrots'] = sides_simplified['carrots'].replace('Carrots', 1)
sides_simplified['cauliflower'] = sides_simplified['cauliflower'].replace('Cauliflower', 1)
sides_simplified['corn'] = sides_simplified['corn'].replace('Corn', 1)
sides_simplified['cornbread'] = sides_simplified['cornbread'].replace('Cornbread', 1)
sides_simplified['salad'] = sides_simplified['salad'].replace('Fruit salad', 1)
sides_simplified['green_beans'] = sides_simplified['green_beans'].replace('Green beans/green bean casserole', 1)
sides_simplified['mac_cheese'] = sides_simplified['mac_cheese'].replace('Macaroni and cheese', 1)
sides_simplified['mashed_potatoes'] = sides_simplified['mashed_potatoes'].replace('Mashed potatoes', 1)
sides_simplified['rolls_biscuits'] = sides_simplified['rolls_biscuits'].replace('Rolls/biscuits',1)
sides_simplified['squash'] = sides_simplified['squash'].replace('Squash',1)
sides_simplified['salad'] = sides_simplified['salad'].replace('Vegetable salad',1)
sides_simplified['yams_sweet_potato'] = sides_simplified['yams_sweet_potato'].replace('Yams/sweet potato casserole', 1)
sides_simplified
```
