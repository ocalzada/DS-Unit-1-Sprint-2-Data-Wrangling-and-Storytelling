```python
# TODO

import pandas as pd

income = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--income_per_person_gdppercapita_ppp_inflation_adjusted--by--geo--time.csv')

lifespan = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--life_expectancy_years--by--geo--time.csv')

population = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--population_total--by--geo--time.csv')

entities = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv')

concepts = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--concepts.csv')

```

```python
#geography, time, adjusted income under 'income' dataframe
income.head()
```

```python
#geography, time and life-expectancy
lifespan.head()
```

```python
#geography, time and population by country
population.head()
```

```python
entities.head()
```

```python
concepts.head()
```

```python
#merge databases

df = pd.merge(income, lifespan)
print(df.shape)
df
```

```python
df = pd.merge(df, population)
print(df.shape)
df.head()
```

```python
entities['world_6region'].value_counts()
```

```python
entities['world_4region'].value_counts()
```

```python
entities_to_keep = ['country', 'name', 'world_4region', 'world_6region']
entities = entities[entities_to_keep]
entities.head()
```

```python
df = pd.merge(df, entities, left_on='geo', right_on='country')
df.head()
```

```python
merged = df.drop('geo', axis='columns')
merged.head()
```

```python
merged = merged.rename(columns={'time':'year',
                               'income_per_person_gdppercapita_ppp_inflation_adjusted': 'income',
                               'life_expectancy_years': 'lifespan',
                               'population_total': 'population',
                               'country': 'country_code',
                               'name':'country',
                               'world_4region': 'region',
                               'world_6region': '6region'})
merged.head()
```

```python
merged.dtypes
```

```python
merged.describe()
```

```python
merged.describe(exclude='number')
```

```python
merged.country.unique()
```

```python
usa = merged[merged.country == 'United States']
usa.head()
```

```python
usa[usa.year.isin([1818, 1918, 2018])]
```

```python
china = merged[merged.country == 'China']
china[china.year.isin([1818,1918, 2018])]
```

```python
import seaborn as sns

now = merged[merged.year==2018]

now.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.relplot(x='income', y='lifespan', hue='6region', size='population', sizes=(30,400), data=now)
plt.xscale('log')
plt.ylim(20,85)
plt.title('The World in 2018')
plt.show()
```

```python
now.sort_values('income', ascending=False)
```

```python
now_qatar = now[now.country=='Qatar']
now_qatar.head()
```

```python
sns.relplot(x='income', y='lifespan', hue='6region', size='population', sizes=(30,400), data=now)
plt.xscale('log')
plt.ylim(20,85)
plt.title('The World in 2018')
plt.text(now_qatar.income, now_qatar.lifespan+.75, s='Qatar')
plt.show()
```

```python
years = [1818,1918,2018]
centuries = merged[merged.year.isin(years)]

fig = sns.relplot(x='income', y='lifespan', hue='6region', size='population', 
                 sizes=(30,400), col='year', data=centuries)

plt.xscale('log')
plt.text(x=now_qatar.income, y=now_qatar.lifespan+.75, s='Qatar')
axes = fig.axes.flatten()
axes[0].set_title(1818)
axes[1].set_title(1918)
axes[2].set_title(2018)

plt.show()
```

```python
years = [1918,1938,1958, 1978, 1998, 2018]
decades = merged[merged.year.isin(years)]
```

```python
sns.relplot(x='income', y='lifespan', hue='6region', size='population', col='year', data=decades)
plt.show();
```

```python
for year in years:
  sns.relplot(x='income', y='lifespan', hue='6region', size='population', 
             data=merged[merged.year==year])
  plt.xscale('log')
  plt.xlim((150,150000))
  plt.ylim((0,90))
  plt.title(year)
  plt.axvline(x=1000, color='grey')
  plt.axhline(y=40, color='grey')
  
  
```
